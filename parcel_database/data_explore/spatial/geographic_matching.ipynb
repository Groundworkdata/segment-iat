# Merging spatial and parcel data
We have an updated dataset for the parcels in Wakefield. We will merge this data with the spatial data so that we can better relate the two databases.

```python
import geopandas as gpd
import pandas as pd
```

```python
FILEPATH = "./database/shape_files/elec_metersPar.shp"
```

```python
gdf = gpd.read_file(FILEPATH)
```

```python
gdf
```

```python
gdf.columns
```

```python
WAKEFIELD_FILEPATH = "./database/wakefield_final_parcel_data.csv"
```

```python
wakefield_parcel_df = pd.read_csv(WAKEFIELD_FILEPATH)
```

```python
wakefield_parcel_df
```

## Investigate "None" LOC IDs
Some "LOC ID"s or "STATE CODE"s are `None`, so we should see how many and ignore those for merging

```python
gdf[gdf["LOC_ID"].isnull()]
```

```python
wakefield_parcel_df[wakefield_parcel_df["STATE CODE"].isnull()]
```

```python
merged_df = pd.merge(
    wakefield_parcel_df[~wakefield_parcel_df["STATE CODE"].isnull()],
    gdf[~gdf["LOC_ID"].isnull()],
    how="inner",
    left_on="STATE CODE",
    right_on="LOC_ID"
)
```

```python
type(merged_df)
```

```python
# Convert back to geodataframe
merged_gdf = gpd.GeoDataFrame(merged_df, geometry=merged_df["geometry"])
```

```python
type(merged_gdf)
```

```python
# Going to write the `merged_df` to a CSV for easier investigation
merged_df.to_csv("./data_explore/merged_parcel_location_data.csv", index=False)
```

```python
# Write out merged shapefile
merged_gdf.to_file("./data_explore/merged_geometry.shp")
```
